```python
import tkinter as tk

class Array_GUI():
    
    def __init__(self, root, box_list, size): # global variables
        self.root = root 
        self.box_list = box_list
        self.size = size
    
    def applyToLabel(self):
        for i in range(self.size.get()): # retrieve global var array size
            element = self.box_list[i].get() # Get value from already appended global array
            ArrayLabel = tk.Label(self.root, text = "Array Element " + str(i) + ": " + element)
            ArrayLabel.pack()

    def boxes(self):
        for i in range(self.size.get()): # retrieve global var array size        
            box = tk.Entry(self.root) # create n entries based on self.size
            box.pack()
            self.box_list.append(box) # append to global array

        ApplytoLabel1 = tk.Button(self.root, text = "Submit To Array", command = self.applyToLabel)
        ApplytoLabel1.pack() # go to applyToLabel function

    def print_(self):
        self.root.title("Looping of entry box")
        self.root.geometry("600x600")

        Array = tk.Frame(self.root)
        Array.pack()

        text1 = tk.Label(Array, text = "Enter the Size of Array:",
                       font="Arial 10 bold", fg = "blue")
        text1.grid(row = 0, column = 0, sticky = "w")

        ArraySize = tk.Entry(Array, textvariable = self.size) # self.size is a global var that submits size of array
        ArraySize.grid(row = 0, column = 1, sticky = "w")

        SizeofArray = tk.Button(Array, text = "Submit", command = self.boxes) # go to the boxes function
        SizeofArray.grid(row = 0, column = 2, sticky = "w")

        self.root.mainloop()

gui = Array_GUI(root = tk.Tk(), box_list = [], size = tk.IntVar())

gui.print_()
```
