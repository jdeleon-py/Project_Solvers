```python
# SCRIPT

if __name__ == '__main__':
    PATH = '/Users/jamesdeleon/Documents/Project_Solvers/sudoku_solver/sudoku.csv'
    df = pd.read_csv(PATH)
    count = 1
    
    for puzzle in df['quizzes'][:10000]:
        arr = []
        for digit in puzzle: arr.append(int(digit))
        arr = np.array(arr).reshape(9,9)
        
        with open('../dev/input_files/input' + str(count) + '.txt', 'w') as write_file:
            for row in arr:
                for digit in row:
                    write_file.write(str(digit))
                write_file.write('\n')
        write_file.close()
        count += 1
```
